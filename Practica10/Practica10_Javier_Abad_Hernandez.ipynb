```python
from sklearn.datasets import fetch_openml
import numpy as np
import pandas as pd
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import MinMaxScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score as ac
import time
from tqdm.notebook import tqdm
```

### Ejercicio 1


```python
X, y = fetch_openml(data_id=1560, return_X_y=True)
scaler = MinMaxScaler()
X=scaler.fit_transform(X)

y.shape, X.shape
```

```python
result = pd.DataFrame(index=range(5,51,5), columns=['Entrena', 'Verifica', 'T_Ejecución'])
result.index.name = 'Neuronas Ocultas'
result
```

```python
for h in result.index:
    print(h)
```

```python
# hay que experimentar con un MLP clasificador usando la validación cruzada habitual (k=10).
K=10
skf=StratifiedKFold(n_splits=K)
```

```python
# Se considerará sólo una capa 
# oculta, cuyo tamaño se hará variar de 5 en 5, entre 5 y 50 nodos.

# h es el numero de neuronas ocultas
for h in result.index:
    r_fold = pd.DataFrame(columns=['Entrena', 'Verifica', 'T. Ejecución']) 
    r_fold.index.name = 'Fold'
    k=0
    mlp = MLPClassifier(hidden_layer_sizes=(h), max_iter=1000)
    for train_index, test_index in tqdm(skf.split(X,y), total=K, unit="fold"):
        comienzo = time.time()
        mlp.fit(X[train_index], y[train_index])
        r_fold.loc[k,:]= [mlp.score(X[train_index], y[train_index]), mlp.score(X[test_index], y[test_index]), time.time()-comienzo]

        #printer fold. Loc[k])
        k+= 1 
    print("Tamaño Capa Oculta: ", h)
    display(r_fold)
    print("MEDIA:\n", r_fold.mean())
    print("\nTiempo Total (seg) =", r_fold.sum(axis=0) [2], "\n\n")
    result.loc[h,:]= [r_fold.mean()[0], r_fold.mean()[1], r_fold.sum(axis=0)[2]]
```

```python
result
```

```python
# Mayor precisión media en el test
result['Verifica'] = result['Verifica'].astype(float)
print("La mayor precisión media está en la capa", result.Verifica.idxmax(), "para Verifica: ",result.Verifica.max())
```

```python
result['T_Ejecución'] = result['T_Ejecución'].astype(float)
# Menor tiempo de ejecución
print("El resultado con menor tiempo de ejecución está en la capa",result.T_Ejecución.idxmin(),"siendo: ",result.T_Ejecución.min())
```
