```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import pandas as pd
import numpy as np
import os
import graphviz
```

```python
data = pd.DataFrame(columns=['A', 'B', 'C', 'D', 'E', 'F'], index=np.arange(64, dtype=int))

for i in range(64):
    b = bin(i)[2:]
    aux=''
    for k in range(6-len(b)):
        aux += '0'
    b = aux + b
    for j in range(6):
        data.iloc[i,j] = b[j]
data
```

```python
data = pd.DataFrame(columns=['A', 'B', 'C', 'D', 'E', 'F'], data=np.array(data).astype(int).astype(bool))
data
```

```python
funcion = ((data.A & data.B) | (~data.C & data.D)) ^ (data.E | ~data.F)
```

```python
data['funcion']=funcion
data
```

```python
data.to_csv('out.csv',index=False)
```

<img src="weka.png">

```python
arbol =DecisionTreeClassifier(criterion='entropy')
arbol.fit(X=data.iloc[:,:-1], y=data.iloc[:,-1])
```

```python
export_graphviz(arbol, out_file="treeDecisionFuncion.dot", feature_names=data.columns[:-1],class_names=True, filled=True)
```
