```python
import numpy as np
from scipy.stats import entropy
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_graphviz, export_text
import re
```

```python
tablaPC = pd.DataFrame(columns = ['Presion','Clase'], index = np.arange(6))
tablaPC.Presion = [40, 48, 60, 72, 80, 90]
tablaPC.Clase = ['+', '+', '-', '-', '-', '+']
tablaPC
```

```python
tablaPC.Clase.value_counts(normalize=True)
```

```python
H = entropy(tablaPC.Clase.value_counts(normalize=True),base=2)
H
```

```python
tablaPC1 = tablaPC
tablaPC1.Presion = np.digitize(tablaPC1.Presion, bins = [(48+60) / 2]) #digitize se carga las expresiones originales, despues hay que volver a añadirlas
tablaPC1
```

```python
#Vamos a calcular la ganancia de información para la primera cota
l1=len(tablaPC1)
G1 = H
for v in tablaPC1.Presion.unique():
    #Mediante el metodo .loc, accedemos a las filas y columnas que indiquemos
    class1 = tablaPC1.loc[tablaPC1.Presion == v]
    G1 -= (len(class1)/len(tablaPC1))*entropy(class1.Clase.value_counts(normalize=True), base=2)
    
print("Ganancia de información de la primera cota es:", G1)
```

```python
#Reponer valores destruidos con digitize
tablaPC.Presion=[40,48,60,72,80,90]
tablaPC
```

```python
tablaPC2 = tablaPC
tablaPC2.Presion = np.digitize(tablaPC2.Presion, bins = [(80+90) / 2]) #digitize se carga las expresiones originales, despues hay que volver a añadirlas
tablaPC2
```

```python
#Vamos a calcular la ganancia de información para la primera cota
l2=len(tablaPC2)
G2 = H
for v in tablaPC2.Presion.unique():
    #Mediante el metodo .loc, accedemos a las filas y columnas que indiquemos
    class2 = tablaPC2.loc[tablaPC2.Presion == v]
    G2 -= (len(class2)/len(tablaPC2))*entropy(class2.Clase.value_counts(normalize=True), base=2)
    
print("Ganancia de información de la primera cota es:", G2)
```

```python
#Reponer valores destruidos con digitize
tablaPC.Presion=[40,48,60,72,80,90]
tablaPC
```

```python
# AHORA VAMOS A COMPROBAR QUE COINCIDA CON EL ALGORTIMO J48
clf=DecisionTreeClassifier(criterion='entropy', max_leaf_nodes=2)
```

```python
clf.fit(np.array(tablaPC.Presion).reshape(-1,1), tablaPC.Clase)
```

```python
txt = export_text(clf, feature_names=['Presion'])
print(txt)
```

```python
export_graphviz(clf, out_file='Presion.dot', class_names=tablaPC.Clase)
```

### Pregunta 3
Esta técnica es más eficiente que la aplicada en la PARTE I porque no requiere discretizar todos los atributos numéricos antes de construir el árbol, sino que lo hace dinámicamente durante el proceso de división, eligiendo el mejor punto de corte para cada atributo según el criterio de ganancia. Es por ello que el árbol del método anterior será menos eficiente y de peor calidad, ya que no se centra en los valores más importantes.

### Pregunta 4

```python
Atrib_Num = [tabla.columns[i] for i in [1,2,15]]
Atrib_Num
```

```python
for atributo in Atrib_Num:
    b = []
    for pos in re.finditer(atributo+' '+'<= ', contenido):
        #print (pos, pos.start(), pos.end())
        #print (atributo+' '+'<= ', pos.start(), pos.end())
        if contenido[pos.end():].find(':') < contenido[pos.end():].find('\n'):
            segmento = contenido[pos.end():].find(':')
        else:
            segmento = contenido[pos.end():].find('\n')
            
        b.append(float(contenido[pos.end():pos.end()+segmento]))
        #print(b[-1])
        b.sort()
    print (atributo, b)
    tabla[atributo] = [chr(65+i) for i in np.digitize(tabla[atributo], bins=b)]
```
